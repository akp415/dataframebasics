#### note on memory handling of dataframes

Creating a copy of your df by assigning it to another name (say df2) can create problems for you. Even when you 
assign to a new variable, it is referencing the same data/indices in memory, which means a manipulation on df or df2 will change the same data in memory. 

This is because, pandas.DataFrame is a wrapper around numpy.ndarry. Creating df2 with df as input, there was a new pd.DataFrame wrapper that was created (thus a different memory address), but the data is exactly the same.

Remedy for this is to use pd.DataFrame.copy. df2 gets its own copy of data that can be manipulated independently. 

df2 = df1.copy()  #use this code

```python
#### Column transformer with pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

numeric_features = ['Age', 'Fare']
categorical_features = ['Embarked', 'Cabin']

#Create a pipeline of both Continuous and Categorical Variables
cnts_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='mean')),('scale', StandardScaler())
])
    

categ_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent'))
])


#Apply Column TRansformers
#List of (name, transformer, column(s)) tuples specifying the transformer objects to be applied to subsets of the data.
#ColumnTransformer ([('name', pipeline_name, features)])


preprocess_pipeline = ColumnTransformer([
    ('continuous', cnts_pipeline, numeric_features),
    ('cat', categ_pipeline, categorical_features)
    ], remainder = 'passthrough') 

df2_processed = preprocess_pipeline.fit_transform(df)
```
